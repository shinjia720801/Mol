### 2-1 List

List 是 Python 最常用的資料結構，特性：
- **有序**：元素按插入順序排列
- **可變**：可以新增、修改、刪除元素
- **可混型**：同一個 list 可以放不同型別

````python
# 建立與基本操作

fruits = ['apple', 'banana', 'cherry']
fruits
fruits[0] # 正向索引
fruits[-1] # 負向索引
fruits[1:3] # 切片
len(fruits) # 看list的長度
````

````python
# 常用方法

fruits.append('date') # 將元素附加到list的尾端
fruits

fruits.insert(1, 'avo') # 將元素插入到指定的index
fruits

fruits.remove('banana') # 將元素進行刪除
fruits

poped = fruits.pop() # 取出並刪除最後一個
print(f'pop: {poped}，剩下:{fruits}')

'apple' in fruits # 判斷元素是否存在於list內
````

````python
# 對list做排序的動作

# 這是一種排序的寫法
nums = [3, 1, 4, 1, 5, 9, 2, 6]
sorted(nums)
sorted(nums, reverse=True)

# 這是另一種排序的寫法
nums.sort()
print(nums)

nums.sort(reverse=True)
print(nums)
````

````python
# List Comprehension

# 產生平方數
squres = [i ** 2 for i in range(1, 6)]
print(f'平方數: {squres}')

# 加上條件篩選
even_squrares = [ i ** 2 for i in range(1, 11) if i % 2 == 0]
print(f'偶數的平方:{even_squrares}')

# 字串處理
words = ['hello', 'world', 'python']
upper_words = [w.upper() for w in words]
print(f'大寫:{upper_words}')
````

#### 🖊️ 隨堂練習 2-1
給定成績 list，用 list comprehension 篩選出 >= 60 的分數，並計算平均。

期望輸出：
```
及格成績：[85, 91, 73, 66, 94, 78]
及格平均：81.2
```

````python
scores = [85, 42, 91, 57, 73, 38, 66, 94, 51, 78]

passed = [s for s in scores if s >= 60]
avg = sum(passed) / len(passed)

print(f'及格成績:{passed}')
print(f'及格平均:{avg:.2f}')
````

### 2-2 Tuple

Tuple 與 List 類似，但**不可變（immutable）**，適合存放不應被修改的資料（如座標、RGB值、函數多回傳值）。

````python
# 基本操作

point = (3, 7)
rgb = (255, 128, 0)

point[0]
rgb[1]
len(rgb)
len(point)

# point.append(5)
# rgb.insert(3, 0.2)

# 拆包(unpacking)
x, y = point
print(f'x={x}, y={y}')
r, g, b = rgb
print(f'R={r}, G={g}, B={b}')
````

````python
# 假設我們要讓函數回傳多個值，那他會用什麼結構去儲存呢? 答案就是tuple

def div(a, b):
  return a // b, a % b # 因為回傳多個值，所以Python會自動打包成一個tuple

quo, rem = div(17, 5) # (3, 2)
print(f'17 / 5 = {quo}..{rem}')
````

````python
# zip:同時可以去訪問多個list，回傳tuple

names = ['Alice', 'Bob', 'Carol']
scores = [92, 78, 85]

for name, score in zip(names, scores):
  print(f'{name}:{score} 分')

for item in zip(names, scores):
  print(item, type(item))
````

### 2-3 Dict

Dict（字典）以 **key-value 對**儲存資料，特性：
- key 必須唯一且不可變（通常用字串或數字）
- value 可以是任何型別，包括 list 或另一個 dict

````python
# 建立dict

student = {'name':'Alice', 'age':20, 'score':[85, 92, 78]}
student['age']
student['score'][1]

# 在dict進行新增，修改和刪除
student['grade'] = 'A'
student
student['age'] = 21
student

del student['age']
student
ren = student.pop('grade', None)
student
ren
````

````python
# 常用方法
grades = {'Alice':92, 'Bob':78, 'Carol':50}
grades
grades.keys() # 列出所有的key
grades.values() # 列出所有的value
grades.items() # 列出所有的key和value
100 in grades
````

````python
# 一次巡覽整個dict
grades

# 正常的寫法
for name, score in grades.items():
  if score >= 60:
    status = '及格'
  else:
    status = '不及格'
  print(f'{name}: {score} 分({status})')

# 簡潔的寫法
for name, score in grades.items():
  status = '及格' if score >= 60 else '不及格'
  print(f'{name}: {score} 分({status})')
````

````python
# Dict Comprehension
grades = {'Alice':92, 'Bob':78, 'Carol':50, 'Dave':55}

# 請你篩選出80分以上的學生
passed = {name: score for name, score in grades.items() if score >= 80}
print(f'80分以上:{passed}')

# 請你將每一個人的分數加5分(最高就是100分)
adj = {name:min(scores + 5, 100) for name, scores in grades.items()}
print(f'調分後:{adj}')
````

## 隨堂練習 2-3｜Dict

**題目：** 統計文字中每個字母出現次數（不含空格，不區分大小寫）。

期望輸出（部分）：`{'h': 1, 'e': 1, 'l': 3, 'o': 2, ...}`

````python
text = "hello world"

# 正常的寫法
freq = {}
for ch in text:
  if ch == ' ':
    continue
  freq[ch] = freq.get(ch, 0) + 1 # 如果字母已存在於字典內，就回傳1，若不存在，就回傳0
print(freq)

# dict comprehension
freq2 = {ch: text.count(ch) for ch in text if ch != ' '}
print(freq2)
````

### 2-4 Set

Set（集合）特性：
- **無序**：不支援索引
- **不重複**：自動去除重複元素
- 支援數學集合運算（聯集、交集、差集）

````python
# 集合運算

a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(f'聯集: {a | b}')
print(f'交集: {a & b}')
print(f'差集: {a - b}')
print(f'對稱差: {a ^ b}')
````

````python
# 最常見的用途: 用來去除重複的元素

names = ['Alice', 'Bob', 'Alice', 'Carol', 'Bob', 'Dave']
print(names)
unique = set(names)
print(unique)
unique = list(unique)
print(unique)
print(f'原始的:{len(names)}筆, 去除重複後:{len(unique)}筆')
````
