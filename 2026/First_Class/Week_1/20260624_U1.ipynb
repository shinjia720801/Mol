#### 🖊️ 隨堂練習 1-4
用 for 迴圈計算 1 到 100 的總和，以及其中所有偶數的總和。

````python
# 方法1:用for迴圈的方式寫

total = 0 # 用來儲存1到100的總和
even_total = 0 # 用來儲存所有偶數的總和

for i in range(1, 101):
  total += i # total = total + i
  # 已經求出1到100的總和
  if i % 2 == 0: # 偶數的判斷
    even_total += i
  # 已經求出1到100的偶數總和
print(f'1到100的總和:{total}')
print(f'偶數總和:{even_total}')


# 方法2:用內建函數驗證
print(f'1到100的總和:{sum(range(1, 101))}')
print(f'偶數總和:{sum(i for i in range(1, 101) if i % 2 == 0)}')
````

````python
# BMI計算機

def bmi(height, weight):
  height = height / 100 # 公分換算成公尺
  value = weight / height ** 2 # 計算BMI公式

  if value < 18.5:
    category = '過輕'
  elif value < 24:
    category = '正常'
  elif value < 27:
    category = '過重'
  else:
    category = '肥胖'
  return round(value, 2), category

bmi_value, cat = bmi(175, 70)
print(f'BMI = {bmi_value}, 分類:{cat}')

test_cases = [(170, 50), (180, 80), (190, 100)]
for h, w in test_cases:
  v, c = bmi(h, w)
  print(f'身高:{h}cm, 體重:{w}Kg, BMI:{v}, 分類:{c}')
````

#### 🖊️ 隨堂練習 1-5
撰寫一個函數 `c_to_f(c)`，將攝氏溫度轉換為華氏溫度。  
公式：`F = C × 9/5 + 32`  
測試：0°C → 32°F，100°C → 212°F，37°C → 98.6°F

````python
def c_to_f(c):
  return c * 9 / 5 + 32

test_cases = [0, 100, 37]
for c in test_cases:
  f = c_to_f(c)
  print(f'攝氏{c:>5}度C, -> 華氏{f:>5}度C') # :>5將變數靠右對齊5個字元
  print(f'攝氏{c:<5}度C, -> 華氏{f:<5}度C') # :<5將變數靠左對齊5個字元
  print(f'攝氏{c:^5}度C, -> 華氏{f:^5}度C') # :^5將變數置中對齊5個字元
````

````python
# 猜數字基本版

import random # 導入random，用來讓遊戲隨機選擇數字

def play_guess():
  answer = random.randint(1, 10) # 從數字1-10之間隨機挑一個數字
  attempts = 0 # 主要是用來記錄猜的次數

  while True:
    guess = int(input('你的猜測:'))
    attempts += 1 # 因為我要記錄猜的次數

    if guess < answer:
      print('太小了，再試其他的數字看看')
    elif guess > answer:
      print('太大了，再試其他數字')
    else:
      print(f'答對了，答案是 {answer}，你總共猜了{attempts}次')
      break

play_guess()
````

````python
# 猜數字再玩一次的版本

import random # 導入random，用來讓遊戲隨機選擇數字

def play_guess():
  while True: # 外層迴圈，用來控制再玩一次的功能
    answer = random.randint(1, 10) # 從數字1-10之間隨機挑一個數字
    attempts = 0 # 主要是用來記錄猜的次數

    while True: # 內層迴圈，用來控制猜測的主程式
      guess = int(input('你的猜測:'))
      attempts += 1 # 因為我要記錄猜的次數

      if guess < answer:
        print('太小了，再試其他的數字看看')
      elif guess > answer:
        print('太大了，再試其他數字')
      else:
        print(f'答對了，答案是 {answer}，你總共猜了{attempts}次')
        break
    again = input('再玩一局?(y/n)')
    if again != 'y':
      print('感謝遊玩，掰掰')
      break

play_guess()
````

````python
# 猜數字判斷是否輸入整數的版本

import random # 導入random，用來讓遊戲隨機選擇數字

def play_guess():
  while True: # 外層迴圈，用來控制再玩一次的功能
    answer = random.randint(1, 10) # 從數字1-10之間隨機挑一個數字
    attempts = 0 # 主要是用來記錄猜的次數

    while True: # 內層迴圈，用來控制猜測的主程式
      raw = input('你的猜測')
      if not raw.isdigit():
        print('請輸入正整數')
        continue
      guess = int(raw)
      attempts += 1 # 因為我要記錄猜的次數

      if guess < answer:
        print('太小了，再試其他的數字看看')
      elif guess > answer:
        print('太大了，再試其他數字')
      else:
        print(f'答對了，答案是 {answer}，你總共猜了{attempts}次')
        break
    again = input('再玩一局?(y/n)')
    if again != 'y':
      print('感謝遊玩，掰掰')
      break

play_guess()
````

````python
# 猜數字限制次數的版本

import random # 導入random，用來讓遊戲隨機選擇數字

def play_guess():
  Max_attempts = 7 # 定義最大猜測次數
  while True: # 外層迴圈，用來控制再玩一次的功能
    answer = random.randint(1, 100) # 從數字1-10之間隨機挑一個數字
    attempts = 0 # 主要是用來記錄猜的次數

    while attempts < Max_attempts: # 內層迴圈，改掉True讓他是有限制的遊玩
      remain = Max_attempts - attempts
      raw = input(f'剩餘{remain} 次，你的猜測:')

      if not raw.isdigit():
        print('請輸入正整數')
        continue

      guess = int(raw)
      attempts += 1 # 因為我要記錄猜的次數

      if guess < answer:
        print('太小了，再試其他的數字看看')
      elif guess > answer:
        print('太大了，再試其他數字')
      else:
        print(f'答對了，答案是 {answer}，你總共猜了{attempts}次')
        break
    else:
      print(f'次數用完!正確答案是 {answer}，下次加油')
    again = input('再玩一局?(y/n)')
    if again != 'y':
      print('感謝遊玩，掰掰')
      break

play_guess()
````
